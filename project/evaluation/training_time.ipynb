```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
neurons = [4, 8, 16, 32, 64, 128]
times = [[301.2, 304.82, 314.56, 326.72, 388.46, 623.84],
        [487.49, 499.91, 516.06, 557.6, 675.94, 1129.96],
        [686.25, 705.11, 722.09, 781.28, 961.83, 1649.99],
        [905.18, 931.16, 972.52, 1029.73, 1275.34, 2121.93]
        ]
legends = ['2 hidden layers', '4 hidden layers', '6 hidden layers','8 hidden layers']

table = [[86.59, 88.82, 92.95, 94.85, 95.60, 96.08],
         [76.48, 87.69, 92.99, 94.90, 95.70, 96.58],
         [64.90, 88.82, 92.72, 95.25, 95.38, 96.14],
         [65.84, 87.42, 91.01, 95.08, 95.83, 95.73]]
acc_128 = [96.08, 96.58, 96.14, 95.73]
offset = np.arange(8, 30, 5)

fig, ax = plt.subplots(figsize=(13,7))
ax2 = ax.twinx()
for i, acc in enumerate(acc_128):
    ax2.bar(128 + offset[i], acc, width=4,label=legends[i])
for i,time in enumerate(times):
    ax.plot(neurons, time, '-o',label=legends[i])

ax.grid(False)
ax2.grid(False)
ax.set_xticklabels(neurons,fontsize=14)
ax.set_yticklabels(np.arange(250, 2500,250),fontsize=14)
ax2.set_yticklabels(np.arange(0, 120,20),fontsize=14)
#plt.xticks(neurons, fontsize=14)
#ax.set_yticks(fontsize=14)
ax.set_ylabel('Training time [s]', fontsize=18)
ax2.set_ylabel('Model avg accuracy [%]', fontsize=18)
ax.set_xlabel('Number of neurons', fontsize=18)
#ax.legend()
ax2.legend(fontsize=15)
plt.savefig('timing_cmp.eps', bbox_inches='tight',format='eps', dpi=500, transparent=True)

plt.show()
```
