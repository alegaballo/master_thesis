```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('train_hist.pkl','rb') as f:
    th = pickle.load(f)
with open('test_metrics.pkl', 'rb') as f:
    tm = pickle.load(f)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid()
plt.title('Model accuracy')
plt.plot(th[0]['categorical_accuracy'])
plt.plot(th[0]['val_categorical_accuracy'],'--')
plt.plot(th[1]['categorical_accuracy'], '-.')
plt.plot(th[1]['val_categorical_accuracy'], ':')
plt.legend(['Training accuracy', 'Validation accuracy','Training accuracy with normalization', 'Validation accuracy with normalization'])
plt.subplot(122)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
plt.title('Model loss')
plt.plot(th[0]['loss'])
plt.plot(th[0]['val_loss'], '--')
plt.plot(th[1]['loss'], '-.')
plt.plot(th[1]['val_loss'], ':')
plt.legend(['Training loss', 'Validation loss','Training loss with normalization', 'Validation loss with normalization'], loc='upper right')
plt.savefig('normalization_cmp_grid.pdf')
plt.show()
```
