```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
with open('train_hist.pkl','rb') as f:
    th = pickle.load(f)
with open('test_metrics.pkl', 'rb') as f:
    tm = pickle.load(f)
```

```python
#sns.set_style('ticks')
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.xlabel('Epoch',fontsize=16)
plt.ylabel('Accuracy',fontsize=16)
plt.grid(False)
plt.title('Model Accuracy',fontsize=16)
plt.plot(th[0]['categorical_accuracy'])
#plt.plot(th[0]['val_categorical_accuracy'],'--')
plt.plot(th[1]['categorical_accuracy'], '-.')
#plt.plot(th[1]['val_categorical_accuracy'], ':')
plt.legend(['Training accuracy', 'Training accuracy with normalization'], fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
#sns.despine()

plt.subplot(122)
plt.xlabel('Epoch',fontsize=16)
plt.ylabel('Loss',fontsize=16)
plt.grid(False)
plt.title('Model loss', fontsize=16)
plt.plot(th[0]['loss'])
#plt.plot(th[0]['val_loss'], '--')
plt.plot(th[1]['loss'], '-.')
#plt.plot(th[1]['val_loss'], ':')
plt.legend(['Training loss', 'Training loss with normalization'], fontsize=14,loc='upper right')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('normalization_cmp.eps', bbox_inches='tight',format='eps', dpi=500, transparent=True)

plt.show()
```
