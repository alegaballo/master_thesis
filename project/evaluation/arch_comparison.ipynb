```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

table = [[88.82, 92.95, 94.85, 95.60, 96.08],
         [87.69, 92.99, 94.90, 95.70, 96.58],
         [88.82, 92.72, 95.25, 95.38, 96.14],
         [87.42, 91.01, 95.08, 95.83, 95.73]]

sns.set_style("ticks")
plt.figure(figsize=(13,5))
ax=sns.heatmap(table, vmin=75, vmax=97, xticklabels=[4,8,16,32,64,128],yticklabels=[2,4,6,8], cmap="PuBu",linewidths=.0, 
            annot=True, fmt='.2f',cbar_kws={'label': 'Accuracy %'})
ax.figure.axes[-1].set_ylabel('Accuracy %', size=16)
plt.yticks(fontsize=14,rotation=0)
plt.xticks(fontsize=14,rotation=0)
plt.xlabel('Number of neurons', fontsize=18)
plt.ylabel('Number of hidden layers', fontsize=18)
plt.savefig('architecture_cmp.eps', bbox_inches='tight',format='eps', dpi=500, transparent=True)
plt.show()
```

```python
xticks=[2,4,6,8]
palette = ['b', 'y', 'r', 'c', 'm', 'g']
labels = ['128 neurons',  '64 neurons', '32 neurons', '16 neurons', '8 neurons', '4 neurons']
plt.figure(figsize=(12,8))
for i, r in enumerate(np.transpose(table)[::-1]):
    ax = sns.barplot(x=[2,4,6,8],y=r, color=palette[i], label=labels[i])
    for bar in ax.patches:
        bar.set_width(0.3)
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(15,7))

for i,x in enumerate(np.transpose(table)):
    plt.plot([2,4,6,8],x, 'o-',label=labels[-1-i])
plt.legend()
plt.show()
```
